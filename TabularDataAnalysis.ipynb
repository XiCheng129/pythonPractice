# Crosstab and Pivot Tables

```python
import transportation_tutorials as tt
import pandas as pd
import numpy as np
```

```python
hh = pd.read_csv(tt.data('SERPM8-BASE2015-HOUSEHOLDS'), index_col=0)
hh.set_index('hh_id', inplace=True)
```

```python
# Count of persons per HH
persons = pd.read_csv(tt.data('SERPM8-BASE2015-PERSONS'))
hh = hh.merge(
    persons.groupby('hh_id').size().rename('hhsize'),
    left_on=['hh_id'],
    right_index=True,
)
```

```python
trips = pd.read_csv(tt.data('SERPM8-BASE2015-TRIPS'))
hh = hh.merge(
    trips.groupby(['hh_id']).size().rename('n_trips'),
    left_on=['hh_id'],
    right_index=True,
)
```

```python
trips
```

```python
hh.pivot_table(
    index='hhsize',
    columns='autos',
    values='n_trips',
)
```

```python
hh.pivot_table(
    columns='autos',
    values='n_trips',
)
```

```python
hh.pivot_table(
    index='hhsize',
    values='n_trips',
)
```

## Marginals

```python
hh.pivot_table(
    index='hhsize',
    columns='autos',
    values='n_trips',
    margins=True,
)
```

## Aggregation Functions

```python
hh.pivot_table(
    index='hhsize',
    columns='autos',
    values='n_trips',
    aggfunc=['mean', 'count', 'sem'],
)
```

```python
hh.pivot_table(
    index='hhsize',
    columns='autos',
    aggfunc='size',
)
```

```python
hh.pivot_table(
    index='hhsize',
    columns='autos',
    aggfunc='size',
    fill_value=0,
)
```

## Using Binning of Continuous Variables

```python
hh.pivot_table(
    index='income',
    columns='autos',
    values='n_trips',
)
```

### For pivot tables, the pandas.qcut function is often a good choice for this, as it will bin the variable so that each bin has a similar total number of observations from the source table.

```python
hh.pivot_table(
    index=pd.qcut(hh.income, 5),
    columns='autos',
    values='n_trips',
)
```

# Linear Regression

```python
import transportation_tutorials as tt
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
hh = pd.read_csv(tt.data('SERPM8-BASE2015-HOUSEHOLDS'), index_col=0)
hh.set_index('hh_id', inplace=True)
```

```python
trips = pd.read_csv(tt.data('SERPM8-BASE2015-TRIPS'))
```

```python
hh.head()
```

```python
trips.head()
```

```python
hh = hh.merge(
    trips.groupby(['hh_id']).size().rename('n_trips'),
    left_on=['hh_id'],
    right_index=True,
)
```

```python
hh.head()
```

```python
hh.info()
```

```python
mod = sm.OLS(
    hh.n_trips,
    sm.add_constant(hh[['autos','income']])
)
res = mod.fit()
print(res.summary())
```

```python
sm.add_constant(hh[['autos','income']]).std()
```

```python
hh['income_100k'] = hh.income / 100_000
```

```python
mod = sm.OLS(
    hh.n_trips,
    sm.add_constant(hh[['autos','income_100k']])
)
res = mod.fit()
print(res.summary())
```

## Piecewise Linear Functions

### The larch package includes a piecewise_expansion function that can expand a single column of data in a pandas.DataFrame into multiple columns based on defined breakpoints.

```python
from larch.util.data_expansion import piecewise_expansion
```

```python
piecewise_expansion(hh.income, [25_000, 75_000]).head()
```

```python
hh.autos
```

```python
hh_edited = pd.concat([
    hh.autos,
    piecewise_expansion(hh.income_100k, [.25, .75]),
], axis=1)

hh_edited.head()
```

### Then we can use this modified dataframe to construct a piecewise linear OLS regression model. Because the original and modified dataframes have the same index (i.e. number and order of rows) we can mix them in the OLS defintion, using the n_trips column from the original as the dependent variable and the explanatory data from the modified dataframe.

```python
mod = sm.OLS(
    hh.n_trips,
    sm.add_constant(hh_edited)
)
res = mod.fit()
print(res.summary())
```

## Polynomial Functions

```python
hh['autos^2'] = hh['autos'] ** 2
hh['income^2'] = hh['income_100k'] ** 2
```

```python
hh.head()
```

```python
mod = sm.OLS(
    hh.n_trips,
    sm.add_constant(hh[['autos','income_100k', 'autos^2', 'income^2']])
)
res = mod.fit()
print(res.summary())
```

```python
def polynomial(x, **kwargs):
    from sklearn.preprocessing import PolynomialFeatures
    poly = PolynomialFeatures(**kwargs)
    arr = poly.fit_transform(x)
    return pd.DataFrame(arr, columns=poly.get_feature_names(x.columns), index=x.index)
```

```python
hh_poly = polynomial(hh[['autos','income_100k']], degree=3)
hh_poly.head()
```

```python
mod = sm.OLS(
    hh.n_trips,
    hh_poly
)
res = mod.fit()
print(res.summary())
```
